```python
# 차 이름 -> model, maker 

name = pd.read_csv('name_to_model(full).csv')
name
```

```python
# total 에다 차 model, maker column 매칭 추가

total = total.merge(name[['name', 'model', 'maker']], left_on = 'car', right_on = 'name', how = 'left')
total
```

```python
# product 번호 - 제품명 매칭

product = pd.read_csv('product.csv')
product = product.rename(columns = {'name' : 'product_name'})  # name column -> product_name
product
```

```python
total_item = total.copy()  # total 복사해서 df 생성

# 1부터 6까지 index에 대해
# %d : 10진수 정수
# "문자열%d" %숫자 -> %d 자리에 % 뒤에 있는 숫자 넣어짐
# product 번호, product name에 대해 merge
# product_1/2/3/4/5/6 in total 에 대해
# product 번호 겹치면 채워 넣어주고..

for i in range(1, 7):
  total_item = total_item.merge(product[['product', 'product_name']], left_on = 'product_%d'%i, right_on = 'product', how = 'left')
  total_item = total_item.rename(columns = {'product_name' : 'product_%d_name' % i})   # product 번호별로 column rename
  total_item = total_item.drop(columns = ['product'])   # product dataFrame에서 가져온 'product' column 삭제

total_item.head()
```

```python
# 앞으로는 이 csv 불러서 바로 사용 -> 위 과정들 생략하자!

total_item.to_csv('total_merged_info.csv', encoding = 'utf-8')
```

# 1. 제품별 가격대

```python
# nunique -> nan은 반영되지 않음

print(total_item.iloc[1][['product_1_name','product_2_name', 'product_3_name', 'product_4_name', 'product_5_name', 'product_6_name']])
total_item.iloc[1][['product_1_name','product_2_name', 'product_3_name', 'product_4_name', 'product_5_name', 'product_6_name']].nunique()
```

한 종류로 시공한 경우로 따져보자.

## 1-1. 1, 2, 4, 5를 한 제품으로 시공할 경우 가격

```python
# 1, 2, 4, 5만 시공한 경우 sorting 한 dataframe -> total_1245

total_1245 = total_item[(total_item['product_3_name'].isnull())&(total_item['product_6_name'].isnull())][['price','maker','product_1_name','product_2_name','product_4_name','product_5_name']]
total_1245 = total_1245.dropna(axis = 0, how = 'any')  # 결측치 제거
total_1245
```

```python
# Dictionary --> 제품명 : [가격]
price_1245 = {}

for i in range(len(total_1245)):
    product = total_1245.iloc[i]['product_1_name']
    price = total_1245.iloc[i]['price']
    
    # 한 가지의 제품명만 있다면
    if total_1245.iloc[i][['product_1_name','product_2_name', 'product_4_name', 'product_5_name']].nunique() == 1:
        if product in price_1245.keys():
            price_1245[product].append(price)
        else:
            price_1245[product] = [price]
price_1245
```

### 평균값

```python
# Dictionary --> 제품명 : 평균 가격

average = {}

for i in price_1245.keys():
    average[i] = np.mean(price_1245[i])
    
print(average)
```

```python
# Average Dataframe

av_df = pd.DataFrame(average.values(), average.keys(), columns = ['price']).sort_values(by = 'price')
av_df.reset_index(inplace = True)
av_df
```

```python
# 1245 시공 제품별 평균 가격

fig= plt.figure(figsize = (20,10))

bar = plt.barh(av_df['index'], av_df['price'], color = 'green', alpha = 0.3)  # 투명도, 색 설정
    
for rect in bar:
        width = rect.get_width()
        plt.text(width + 5000, rect.get_y() + 0.4, '%d' %width, ha = 'left', va = 'center', size = 10)
plt.title('1,2,4,5 조합 단일 제품 시공 평균 가격 지출', fontsize = 20)
plt.xlabel('price(백만원)', fontsize = 15)
plt.ylabel('product', fontsize = 15)
plt.show()
```

```python
# /4 시켜준 값 (4 군데 시공해서)

av_df['div4'] = av_df['price']/4
av_df
```

```python
# 비교용 /4 값 추출 dataframe

av_df[['index', 'div4']]
```

```python
# 1245 시공 제품별 평균 가격

fig= plt.figure(figsize = (20,10))

bar = plt.barh(av_df['index'], av_df['div4'], color = 'green', alpha = 0.3)  # 투명도, 색 설정
    
for rect in bar:
        width = rect.get_width()
        plt.text(width + 2000, rect.get_y() + 0.4, '%d' %width, ha = 'left', va = 'center', size = 10)
plt.title('1,2,4,5 조합 단일 제품 시공 평균 가격 지출(/4)', fontsize = 20)
plt.xlabel('price(\)', fontsize = 15)
plt.ylabel('product', fontsize = 15)
plt.show()
```

## 1-2. 5만 시공할 경우 가격

```python
# 5 하는 경우 제품별 가격대
# 1, 2, 3, 4, 6 null 값인 행들만 추출

total_5 = total_item[(total_item['product_1_name'].isnull())&(total_item['product_2_name'].isnull())&(total_item['product_3_name'].isnull())&(total_item['product_4_name'].isnull())&(total_item['product_6_name'].isnull())][['price','maker','product_5_name']]
total_5 = total_5.dropna(axis = 0, how = 'any')  # 결측치 제거
total_5
```

```python
# 제품명 : [가격들] dictionary 생성

price_5 = {}

for i in range(len(total_5)):
    product = total_5['product_5_name'].iloc[i]
    price = int(total_5['price'].iloc[i])
    if product in price_5.keys():
        price_5[product] = price_5[product] + [price]
    else:
        price_5[product] = [price]

price_5
```

### 평균값

```python
# 제품별 가격 평균 dictionary
# 제품명 : 평균

average_5 = {}

for i in price_5.keys():
    average_5[i] = np.mean(price_5[i])

print(average_5)
```

```python
# 평균가로 datarframe 만들기

av5_df = pd.DataFrame(average_5.values(), average_5.keys(), columns = ['price']).sort_values(by = 'price')
av5_df.reset_index(inplace = True)
av5_df
```

```python
av_df.sort_values(by= 'price',ascending = False)
```

```python
# 5 시공 제품별 평균 가격

fig= plt.figure(figsize = (20,10))

bar = plt.barh(av5_df['index'], av5_df['price'], color = 'green', alpha = 0.3)  # 투명도, 색 설정
    
for rect in bar:
        width = rect.get_width()
        plt.text(width + 1000, rect.get_y() + 0.4, '%d' %width, ha = 'left', va = 'center', size = 10)
plt.title('5 시공 제품별 평균 가격 지출', fontsize = 20)
plt.xlabel('price', fontsize = 15)
plt.ylabel('product', fontsize = 15)
plt.show()
```

### Box plot

```python
# box plot -> list 형태로 데이터 입력해야

price_5_v = []

for i in av5_df['index']:
    price_5_v.append(price_5[i])
    
price_5_v
```

```python
plt.figure(figsize = (20,15))
plt.boxplot(price_5_v, labels = av5_df['index'], showmeans = True, vert = 0)
plt.title('5 시공 제품별 가격 지출 boxplot', fontsize = 20)
plt.xlabel('price(천만원)', fontsize = 15)
plt.ylabel('Product', fontsize = 15)
plt.show()
```

# 2. 브랜드별 Product 선호도

```python
# 브랜드별 1열 위치 선호 제품 그래프 나오는 함수 만들기

def brand_1 (brand, num):   # num 0 : 제품명, num 1 : 제품 번호
    # 해당 브랜드 1번만 추출
    df_1 = total_item[total_item['maker'] == brand][['product_1', 'product_1_name']]
    
    # 결측치 제거
    df_1.dropna(inplace = True)
    
    # 제품 명으로
    if num == 0:  
        df_1_count = df_1.groupby(by = ['product_1_name'], as_index = False).count().sort_values(by = 'product_1')
        
        # 시각화
        fig= plt.figure(figsize = (20,10))

        bar = plt.barh(df_1_count['product_1_name'], df_1_count['product_1'], color = 'orange', alpha = 0.5)  # 투명도, 색 설정
    
        for rect in bar:
            width = rect.get_width()
            plt.text(width + 1, rect.get_y() + 0.4, '%d' %width, ha = 'left', va = 'center', size = 10)
        plt.title(brand + ' 1열 선호 제품 (제품명)', fontsize = 20)
        plt.xlabel('시공수', fontsize = 15)
        plt.ylabel('product', fontsize = 15)
        plt.show()
    
    # 제품 번호로 - top 30
    elif num == 1:
        df_1_count = df_1.groupby(by = ['product_1'], as_index = False).count().sort_values(by = 'product_1_name')
        
        # 리스트 내 요소들 - int --> str 
        df_1_count['product_1'] = list(map(str, df_1_count['product_1']))
        
        # top 30만 자름
        df_1_count = df_1_count.iloc[-30:]
        
        # 시각화
        fig= plt.figure(figsize = (20,10))

        bar = plt.barh(df_1_count['product_1'], df_1_count['product_1_name'], color = 'violet', alpha = 0.5)  # 투명도, 색 설정
    
        for rect in bar:
            width = rect.get_width()
            plt.text(width + 1, rect.get_y() + 0.4, '%d' %width, ha = 'left', va = 'center', size = 10)
        plt.title(brand + ' 1열 선호 제품 top30 (제품 번호)', fontsize = 20)
        plt.xlabel('시공수', fontsize = 15)
        plt.ylabel('product', fontsize = 15)
        plt.show()
```
