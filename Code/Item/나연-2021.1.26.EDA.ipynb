```python
cars_final = pd.read_csv('cars_final.csv')
cars_final
```

```python
total = total.merge(cars_final[['name', 'model', 'maker', 'kind']], left_on = 'car', right_on = 'name', how = 'left')
total
```

```python
product = pd.read_csv('product.csv')
product = product.rename(columns = {'name' : 'product_name'})
product
```

```python
total_item = total.copy()  # total 복사해서 df 생성

# 1부터 6까지 index에 대해
# %d : 10진수 정수
# "문자열%d" %숫자 -> %d 자리에 % 뒤에 있는 숫자 넣어짐
# product 번호, product name에 대해 merge
# product_1/2/3/4/5/6 in total 에 대해
# product 번호 겹치면 채워 넣어주고..

for i in range(1, 7):
  total_item = total_item.merge(product[['product', 'product_name']], left_on = 'product_%d'%i, right_on = 'product', how = 'left')
  total_item = total_item.rename(columns = {'product_name' : 'product_%d_name' % i})   # product 번호별로 column rename
  total_item = total_item.drop(columns = ['product'])   # product dataFrame에서 가져온 'product' column 삭제

total_item.head()
```

# 1. 차종에 따른 (suv, sedan, van 등) 가격 비교

## 1-1. (1, 2, 4, 5) 시공한 경우 차종에 따른 가격

```python
# 1, 2, 4, 5만 시공한 경우 sorting 한 dataframe -> total_1245

total_1245 = total_item[(total_item['product_3_name'].isnull())&(total_item['product_6_name'].isnull())][['price','maker','kind','product_1_name','product_2_name','product_4_name','product_5_name']]
total_1245 = total_1245.dropna(axis = 0, how = 'any')  # 결측치 제거
total_1245
```

### 평균값

```python
# 정직하게 평균 구했을 때
# 브랜드, 제품 등 고려하지 않고

kind_av = total_1245.groupby(by = ['kind'],as_index = False).mean().sort_values(by = 'price')
kind_av
```

```python
# 1245 시공
# 차종별 평균 가격

fig= plt.figure(figsize = (12,5))

bar = plt.barh(kind_av['kind'], kind_av['price'], color = 'blue', alpha = 0.3)  # 투명도, 색 설정
    
for rect in bar:
        width = rect.get_width()
        plt.text(width + 1000, rect.get_y() + 0.3, '%d' %width, ha = 'left', va = 'center', size = 9)
plt.title('1,2,4,5 조합 시공 차종별 평균 가격 지출', fontsize = 20)
plt.xlabel('price(\)', fontsize = 15)
plt.ylabel('Kind', fontsize = 15)
plt.show()
```

van, Three row truck은 규모 때문에 비싼 가격으로 추정. 크기와 상관관계 보이지 않는 Coupe. -> 주로 고급형 모델이 많아서일까?

```python
# /4 한 가격

kind_av['div4'] = kind_av['price']/4
kind_av
```

```python
# 1245 시공 제품별 평균 가격 /4

fig= plt.figure(figsize = (12,5))

bar = plt.barh(kind_av['kind'], kind_av['div4'], color = 'blue', alpha = 0.3)  # 투명도, 색 설정
    
for rect in bar:
        width = rect.get_width()
        plt.text(width + 1000, rect.get_y() + 0.4, '%d' %width, ha = 'left', va = 'center', size = 10)
plt.title('1,2,4,5 조합 차종별 시공 평균 가격 지출(/4)', fontsize = 20)
plt.xlabel('price(\)', fontsize = 15)
plt.ylabel('product', fontsize = 15)
plt.show()
```

### Box Plot

```python
# Dictionary --> 제품명 : [가격]
price_1245 = {}

for i in range(len(total_1245)):
    kind = total_1245.iloc[i]['kind']
    price = total_1245.iloc[i]['price']

    if kind in price_1245.keys():
        price_1245[kind].append(price)
    else:
        price_1245[kind] = [price]
        
price_1245
```

```python
price_1245_v = []

for i in kind_av['kind']:
    price_1245_v.append(price_1245[i])
    
price_1245_v
```

```python
plt.figure(figsize = (15,8))
plt.boxplot(price_1245_v, labels = kind_av['kind'], showmeans = True, vert = 0)
plt.title('1,2,4,5 조합 시공 차종별 가격 지출 boxplot', fontsize = 20)
plt.xlabel('price(천만원)', fontsize = 15)
plt.ylabel('kind', fontsize = 15)
plt.show()
```

# 2. 차종별 선호 제품

## 2-1. 차종별 1열 위치 선호 제품

```python
total_item['kind'].unique()
```

'Mid-Large suv', 'Sedan', 'van', 'Small suv', nan, 'One row truck',
       'Coupe', 'Three row truck', 'Two row truck' 차종 입력

```python
# 차종별 1열 위치 선호 제품 그래프 나오는 함수 만들기

def kind_1 (kind, num):   # num 0 : 제품명, num 1 : 제품 번호
    # 해당 차종 1번만 추출
    df_1 = total_item[total_item['kind'] == kind][['product_1', 'product_1_name']]
    
    # 결측치 제거
    df_1.dropna(inplace = True)
    
    # 제품 명으로
    if num == 0:  
        df_1_count = df_1.groupby(by = ['product_1_name'], as_index = False).count().sort_values(by = 'product_1')
        
        # 시각화
        fig= plt.figure(figsize = (20,10))

        bar = plt.barh(df_1_count['product_1_name'], df_1_count['product_1'], color = 'orange', alpha = 0.5)  # 투명도, 색 설정
    
        for rect in bar:
            width = rect.get_width()
            plt.text(width + 1, rect.get_y() + 0.4, '%d' %width, ha = 'left', va = 'center', size = 10)
        plt.title(kind + ' 1열 선호 제품 (제품명)', fontsize = 20)
        plt.xlabel('시공수', fontsize = 15)
        plt.ylabel('product', fontsize = 15)
        plt.show()
    
    # 제품 번호로 - top 30
    elif num == 1:
        df_1_count = df_1.groupby(by = ['product_1'], as_index = False).count().sort_values(by = 'product_1_name')
        
        # 리스트 내 요소들 - int --> str 
        df_1_count['product_1'] = list(map(str, df_1_count['product_1']))
        
        # top 30만 자름
        df_1_count = df_1_count.iloc[-30:]
        
        # 시각화
        fig= plt.figure(figsize = (20,10))

        bar = plt.barh(df_1_count['product_1'], df_1_count['product_1_name'], color = 'violet', alpha = 0.5)  # 투명도, 색 설정
    
        for rect in bar:
            width = rect.get_width()
            plt.text(width + 1, rect.get_y() + 0.4, '%d' %width, ha = 'left', va = 'center', size = 10)
        plt.title(kind + ' 1열 선호 제품 top30 (제품 번호)', fontsize = 20)
        plt.xlabel('시공수', fontsize = 15)
        plt.ylabel('product', fontsize = 15)
        plt.show()
```
