#### 지역별로 가게는 몇개 씩 있고, 가장 많이 시공을 한 시공왕은 어디일까요오?

```python
import re
from tqdm import tqdm
region_dic = {'경기도' : 0, '전라북도' : 0, '전라남도' : 0, '경상북도' : 0 , '경상남도' : 0, '충청북도' : 0, '충청남도' : 0,
             '서울특별시' : 0, '인천광역시' : 0, '대전광역시' : 0, '광주광역시' : 0, '대구광역시' : 0, '울산광역시' : 0,
             '부산광역시' : 0, '제주특별자치도' : 0, '세종특별자치시' : 0, '강원도' : 0, '기타' : 0}
store_dic = {}

for store_id, store_region in tqdm(zip(minki_df.store, minki_df.new_store)) :
    region = re.findall(r'(\w+)_\d+', store_region)[0]
    
    if store_id in store_dic :
        store_dic[store_id] += 1
    else :
        store_dic[store_id] = 1
        
        try :
            region_dic[region] += 1
        except :
            region_dic['기타'] += 1
        
```

```python
region_dic
```

! 경기도가 가장 많다는 것을 알수가 있슴니다. 세종특별자치시는 애초에 좁은 범위이니 빼면, 강원도랑 제주도가 좀 적은것 같네요

```python
from tqdm import tqdm
tqdm.pandas()
minki_df['works'] = minki_df.store.progress_apply(lambda store : store_dic.get(store))
```

#### 각 지역별 시공왕을 알아봅시다

```python
minki_df['region'] = minki_df.new_store.apply(lambda store : re.findall(r'(\w+)_\d+', store)[0])
```

```python
def region_store_king(region) :
    df_use = minki_df[minki_df.region == region]
    region_total_works = len(df_use)
    region_total_stores = len(df_use.new_store.unique())
    region_max_store = df_use.sort_values('works', ascending = False).store.iloc[0]
    region_max_store_works = len(df_use[df_use.store == region_max_store])
    
    
    print(f"{region}에서는 지금까지 {region_total_stores} 개의 가게들에 의해 총 {region_total_works} 번의 시공이 이루어졌습니다.")
    print(f"{region}에서 시공왕은 {region_max_store} 입니다. 이 가게에서만 {region_max_store_works} 번의 시공이 이루어졌습니다")
    print(f"시공왕의 지역 점유율은 약 {(region_max_store_works*100 / region_total_works):.2f} % 입니다\n")
    
    return(region_max_store)
```

```python
for region in region_dic :
    region_store_king(region)
```

```python
# 각 지역별 Top 3가 차지하는 점유율은 얼마나 될까?
def region_store_top3(region) :
    df_use = minki_df[minki_df.region == region]
    region_total_works = len(df_use)
    region_total_stores = len(df_use.new_store.unique())
    region_max_store_list = list(df_use.sort_values('works', ascending = False).drop_duplicates('new_store').store.iloc[0:3])
    region_max_store_works = [ len(df_use[df_use.store == store]) for store in region_max_store_list ]
    
    region_max_store_dealer_works = 0
    for store in region_max_store_list :
        region_max_store_dealer_works += sum( df_use[df_use.store == store].by_dealer == 1 )
    
    print(f"{region}에서는 지금까지 {region_total_stores} 개의 가게들에 의해 총 {region_total_works} 번의 시공이 이루어졌습니다.")
    print(f"{region}에서 Top 3 가게에 의해서 {sum(region_max_store_works)} 번의 시공이 이루어졌습니다")
    print(f"Top 3의 지역 점유율은 약 {(sum(region_max_store_works)*100 / region_total_works):.2f} % 입니다")
    print(f"Top 3의 총 시공 대비 딜러 시공의 비율은 {(region_max_store_dealer_works *100 / sum(region_max_store_works)):.2f} % 입니다 \n")
    
    return region_max_store_list
```

```python
for region in region_dic :
    try :
        region_store_top3(region)
    except :
        continue
```

```python
def car_region_compare(car, region) :
    df_use = minki_df.loc[(minki_df.model == car) & (minki_df.region == region)].copy()
    max_store = df_use.sort_values('works', ascending = False).store.iloc[0]
    
    result_df = pd.DataFrame(df_use.sort_values('works', ascending = False)[['price', 'new_store']].groupby("new_store").describe())
    result_df = result_df.sort_values([('price', 'count')], ascending = False)
    return result_df
    
```

```python
def car_region_compare_2(car, region, df) :
    if region == 'all' :
        df_use = df.loc[(df.model == car)].copy()
    elif car == 'all' :
        df_use = df.loc[(df.region == region)].copy()
    else :
        df_use = df.loc[(df.model == car) & (df.region == region)].copy()
    max_store = df_use.sort_values('works', ascending = False).store.iloc[0]
    
    result_df = pd.DataFrame(df_use.sort_values('works', ascending = False)[['price', 'new_store']].groupby("new_store").describe())
    result_df = result_df.sort_values([('price', 'count')], ascending = False)
    return result_df, df_use
    
```
